```python
columns_to_keep = [
    'id', 'boxer_name' ,'boxer_slug', 'boxer_boxrec_id', 
    'boxer_nationality', 'boxer_num_wins', 'boxer_num_losses', 'boxer_draws',
    'sex', 'bouts', 'career', 'ko_percent', 'rounds', 'debut', 'residence', 'status'
]

# Keep only the selected columns in the DataFrame
box_two_df = box_df[columns_to_keep]
```

```python
box_two_df['boxer_slug'] = box_two_df['boxer_slug'].str.strip()
box_two_df['boxer_name'] = box_two_df['boxer_name'].str.strip()
box_two_df['career'] = box_two_df['career'].str.strip()
box_two_df['boxer_nationality'] = box_two_df['boxer_nationality'].str.strip('[]').str.strip('"').str.strip("'").str.strip()
box_two_df['ko_percent'] = box_two_df['ko_percent'].str.strip("%").str.strip()
box_two_df['residence'] = box_two_df['residence'].str.strip()

box_two_df
```

**id:**  unique identifier assigned to each boxer in the dataset.

**boxer_name:** The full name of the boxer, used for identification and display.

**boxer_slug::** A URL-friendly version of the boxer's name, typically used as a unique identifier in web applications.

**boxer_boxrec_id:** The unique ID assigned to the boxer on BoxRec, a popular boxing database and record-keeping website.

**boxer_nationality:** The nationality of the boxer, representing the country or countries they are affiliated with.

**boxer_num_wins:** The total number of matches the boxer has won in their professional career.

**boxer_num_losses:** The total number of matches the boxer has lost in their professional career.

**boxer_draws:** The total number of matches the boxer has fought that ended in a draw (no winner was declared).

**sex:** The gender of the boxer, which is either "male" or "female."

**bouts:** The total number of fights (matches) the boxer has participated in, including wins, losses, and draws.

**career:** The range of years the boxer has been active in their professional career, shown as a time span (e.g., 2005-2022).

**ko_percent:** The percentage of wins achieved by knockout (KO). A KO occurs when one boxer is unable to continue due to a decisive hit or injury.

**rounds:**
The total number of rounds fought by the boxer in their career, representing their overall in-ring experience.

**debut:** The date of the boxer‚Äôs first professional fight, marking the beginning of their career.

**residence:**
The location where the boxer resides, which can sometimes differ from their nationality.

**status:**
Information about the boxer‚Äôs current professional status. This might indicate whether they are active, retired, or inactive. If the status is NaN, it may be missing or unrecorded.

```python
box_two_df[['first_name', 'last_name']] = box_two_df['boxer_name'].str.split(' ', n=1, expand=True)
```

```python
box_two_df
```

```python


# Reorder columns to move 'first_name' and 'last_name' to the beginning
box_two_df = box_two_df.rename(columns={'id': 'unique_id'})
```

```python
columns = ['unique_id', 'first_name', 'last_name'] + [col for col in box_two_df.columns if col not in ['unique_id', 'first_name', 'last_name']]
box_two_df = box_two_df[columns]

box_two_df.drop(columns=['boxer_name'], inplace=True)

box_two_df
```

```python
from splink.exploratory import profile_columns

profile_columns(box_two_df, db_api=DuckDBAPI(), top_n=10, bottom_n=5)
```

```python
from splink import block_on
from splink.blocking_analysis import (
    cumulative_comparisons_to_be_scored_from_blocking_rules_chart,
)

blocking_rules_for_analysis = [
    block_on("last_name"),
    block_on("first_name", "residence"),

]

cumulative_comparisons_to_be_scored_from_blocking_rules_chart(
    table_or_tables=box_two_df,
    blocking_rules=blocking_rules_for_analysis,
    db_api=db_api,
    link_type="dedupe_only",
)
```

```python
from splink import Linker, SettingsCreator,  block_on, DuckDBAPI,  comparison_library as cl

settings = SettingsCreator(
    link_type="dedupe_only",
    comparisons=[
        cl.NameComparison("first_name"),
        cl.NameComparison("last_name"),
        cl.ExactMatch("boxer_nationality").configure(term_frequency_adjustments=True),
        cl.ExactMatch("bouts"),
        cl.ExactMatch("career"),
        cl.LevenshteinAtThresholds("residence", [1, 3]),
    ],
    blocking_rules_to_generate_predictions=[
        block_on("last_name"),
        block_on("first_name", "residence"),

    ],
    retain_intermediate_calculation_columns=True,
)

linker = Linker(box_two_df, settings, db_api=DuckDBAPI())
```

```python
deterministic_rules = [
    "l.first_name = r.first_name",
    "l.unique_id = r.unique_id",
]

linker.training.estimate_probability_two_random_records_match(deterministic_rules, recall=0.8)
linker.training.estimate_u_using_random_sampling(max_pairs=1e6, seed=5)

linker.training.estimate_parameters_using_expectation_maximisation(
    block_on("first_name")
)
linker.training.estimate_parameters_using_expectation_maximisation(
    block_on("last_name")
)

# Step 5: Run predictions
results = linker.inference.predict(threshold_match_probability=0.9)

# Step 6: Cluster duplicates
clusters = linker.clustering.cluster_pairwise_predictions_at_threshold(
    results, threshold_match_probability=0.9
)

# Step 7: Save deduplicated results
df_clusters = clusters.as_pandas_dataframe()
```

```python
# Find duplicate cluster_ids
duplicate_clusters = df_clusters[df_clusters.duplicated(subset=['cluster_id'], keep=False)]

# Sort the duplicate records by cluster_id in increasing order
duplicate_clusters_sorted = duplicate_clusters.sort_values(by='cluster_id')

# Display the sorted duplicate records
duplicate_clusters_sorted
```
